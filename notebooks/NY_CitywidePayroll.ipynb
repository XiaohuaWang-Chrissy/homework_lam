```python
import pandas as pd
```

```python
nyc_payroll = pd.read_csv("../data/Citywide_Payroll_Data_(Fiscal_Year)_20260415.csv", low_memory=False)
```

```python
nyc_payroll.columns
```

```python
nyc_payroll.head()
```

```python
nyc_payroll_2025 = nyc_payroll[nyc_payroll['Fiscal Year'] == 2025]
```

```python
nyc_payroll_2025_Active = nyc_payroll_2025[nyc_payroll_2025['Leave Status as of June 30'] == 'ACTIVE']
```

I use Ai for this code below, I don't know how to clean the "$" and "," in the money cell

```python
nyc_payroll_2025_Active['Total OT Paid Clean'] = nyc_payroll_2025_Active['Total OT Paid'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```

```python
agency_ot = (
    nyc_payroll_2025_Active
    .groupby('Agency Name')['Total OT Paid Clean']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)
agency_ot
```

```python
nypd = nyc_payroll_2025_Active[nyc_payroll_2025_Active['Agency Name'] == 'POLICE DEPARTMENT']
```

```python
nypd_title = (
    nypd.groupby('Title Description')['Total OT Paid Clean']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)
nypd_title
```
